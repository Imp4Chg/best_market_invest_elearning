## Exploring the data to find amount spent per month by new coders

The courses offered by Free Code Camp are accessed by subscriptions of USD59 per month. It is therefore essential to identify which location has students who are willing to pay as much to learn web and mobile development.

The column _MoneyForLearning_ contains the total amount spent per student for learning while the _MonthsProgramming_ column has the duration in months they've been learning for.

Dividing total amount spent for learning since survey started **MoneyforLearning** by the number of months spent learning **MonthsProgramming**

__Only USA, India, UK and Canada would be examined since they have significantly higher proportion of coders interested in e-learning and they are all English speaking countries primarily__

```python
# Replacing month values stating '0' with '1' 
# because new coders with month '0' would have been on their first month at least for 'MonthsProgramming'

interested_new_coders['MonthsProgramming'] = interested_new_coders['MonthsProgramming'].copy().replace({0:1})
```

```python
# Dividing total amount spent for learning since survey started MoneyforLearning by the 
# number of months spent learning MonthsProgramming

interested_new_coders['AmountPerMonth'] = \
                        interested_new_coders['MoneyForLearning']/interested_new_coders['MonthsProgramming']
```

```python
# Non-valid amount values for "AmountPerMonth"

interested_new_coders['AmountPerMonth'].isnull().sum()
```

```python
# Selecting only data with valid "AmountPerMonth"

interested_new_coders = interested_new_coders[interested_new_coders['AmountPerMonth'].notnull()]
```

```python
# Selecting only data where Country ('CountryLive' column) is not null

interested_new_coders = interested_new_coders[interested_new_coders['CountryLive'].notnull()]
```

```python
# Getting the average (mean) amount spent per month by new coders per country
av_country = interested_new_coders.groupby('CountryLive').mean()

# Mean amount spent per month by new coders per country for USA, Canada, UK and India (being the top 4 by location)
av_country['AmountPerMonth'][['United States of America','India','United Kingdom','Canada']]
```

The most amount per month is spent in the USA, hinting that it is a good market for advertising the e-learning courses, given that it also has the largest proportion of new coders. The UK and Canada, especially the UK seem have considerably less amount per month on average than India. Given their developed status, one would have expected the average amount spent on e-learning in both countries to be more than that of India.

More exploration of the data may be needed to determine if extreme outliers are responsible for the average figures calculated.

```python
# Median amount spent per month by new coders per country for the top 4 by location

av2_country = interested_new_coders.groupby('CountryLive').median()
av2_country['AmountPerMonth'][['United States of America','India','United Kingdom','Canada']]
```

* The mode does not look right. Zero is returned as the frequency of each of the four countries for the amount spent per month. 
* The median values also don't look representative of the amount of money spent on average. Canada, UK and India all have value of 0.000 for their median, while the US has a median of 3.3 USD. This may indicate the presence of outliers which should be investigated.
* The mean would be the most appropriate average value to use in this scenario

```python
# Extracting dataframe with the top 4 locations of interested coders and displaying first 5 rows

top_4 = interested_new_coders[interested_new_coders['CountryLive'].isin\
                              (['United States of America','United Kingdom','Canada','India'])]
top_4[:5]
```

## Plotting the amount spent per month for the top 4 countries to identify outliers

```python
sns.boxplot(y='AmountPerMonth', x='CountryLive', data=top_4)
plt.title('Amount spent per month \nin top 4 countries', fontsize=16)
plt.ylabel('Amount per month, USD', fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.xticks(range(4),['US','UK','India','Canada'])
plt.show()
```

Some extreme values are visible immediately. A few interested coders in the US seem to spend 50,000 USD and 80,000 USD per month on e-learning. While that is not impossible, it is highly unlikely. Most interested coders seem to spend below 20,000 USD in the US.

10,000 USD per month seems to be an outlier value for interested coders in India, but intially we would set 20,000 USD as the maximum amount that can be spent per month across the top 4 countries.

```python
# Extracting data for interested coders who spend less than 20,000 USD a month only across the top 4 locations

top_4_rev = top_4[top_4['AmountPerMonth'] < 20000]
```

## Plotting the top 4 countries showing interested coders spending under 20,000 USD per month

```python
sns.boxplot(y='AmountPerMonth', x='CountryLive', data=top_4_rev)
plt.title('Amount spent per month \nin top 4 countries', fontsize=16)
plt.ylabel('Amount per month, USD', fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.xticks(range(4),['US','UK','India','Canada'])
plt.show()
```

The box plot above shows the US with the highest number and variation of outliers. India ranks next with the 2nd highest outlier occurence, while Canada seems to have just one outlier.

Examining the outliers would indicate whether the data they hold is valid or not. Data can be termed valid if the interested coders attented quite expensive learning programmes like bootcamps etc.

Extreme outliers in:
* the US - those who paid over USD 6,000 per month for e-learning
* India - those who paid over USD 2,500 per month for e-learning
* Canada - one person who paid USD 5,000 per month for e-learning

### Examining US outliers

```python
us_outliers = top_4_rev[(top_4_rev['AmountPerMonth'] >= 6000) & \
                        (top_4_rev['CountryLive'] == 'United States of America')]
us_outliers
```

Of note from the data for extreme outliers in the US is that:
* there are 11 interested coders who allegedly paid over USD 6,000 per month for e-learning
* 5 of them did not attend any bootcamp, while the other 6 attended 1 each.
* Only 3 of the 11 had spent over 3 months programming

We can assume that those who had spent 3 months and less programming had probably enrolled in a bootcamp that was still on going at the time of the survey and would end up learning for more months without paying extra, reducing their amount spent per month to less than what they had provided.

Consequently, we would remove data for those who did not attend any bootcamp (there's no clear indication as to how they might have spent that much on e-learning courses) and those who have programmed for 3 months or less.

```python
# Removing interested coders who did not attend any bootcamps

no_bootcamp = top_4_rev[(top_4_rev['AmountPerMonth'] >= 6000) & \
                        (top_4_rev['CountryLive'] == 'United States of America') & \
                        (top_4_rev['AttendedBootcamp'] == 0)]

top_4_rev = top_4_rev.drop(no_bootcamp.index)

# Removing interested coders who had spent 3 months or less programming

less_than_3 = top_4_rev[(top_4_rev['AmountPerMonth'] >= 6000) & \
                        (top_4_rev['CountryLive'] == 'United States of America') & \
                        (top_4_rev['MonthsProgramming'] <= 3)]

top_4_rev = top_4_rev.drop(less_than_3.index)
```

### Examining India outliers

```python
india_outliers = top_4_rev[
    (top_4_rev['AmountPerMonth'] >= 2500) &
    (top_4_rev['CountryLive'] == 'India')
]

india_outliers
```

None of the respondents here seem to have attended any bootcamps, so, we would remove data for those who did not attend any bootcamp (there's no clear indication as to how they might have spent that much on e-learning courses)

```python
# Removing interested coders who did not attend any bootcamps in India

top_4_rev = top_4_rev.drop(india_outliers.index)
```

### Examining Canada outlier

```python
canada_outlier = top_4_rev[
    (top_4_rev['AmountPerMonth'] >= 5000) &
    (top_4_rev['CountryLive'] == 'Canada')
]

canada_outlier
```

This respondent from Canada has attendend a bootcamp, but has only been programming for 2 months. Perhaps, the boot camp enrolled in spans longer than 2 months also reducing his amount spent per month.

In that case, we will also remove this extreme outlier.

```python
# Removing interested coders who did not attend any bootcamps in Canada

top_4_rev = top_4_rev.drop(canada_outlier.index)
```

### Calculating the average amount spent per month in each of the top 4 countries by interested coders

```python
top_4_rev.groupby('CountryLive').mean()['AmountPerMonth']
```

### Re-plotting the average amount spent per month in each of the top 4 countries

```python
sns.boxplot(y='AmountPerMonth', x='CountryLive', data=top_4_rev)
plt.title('Amount spent per month \nin top 4 countries', fontsize=16)
plt.ylabel('Amount per month, USD', fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.xticks(range(4),['US','UK','India','Canada'])
plt.show()
```
